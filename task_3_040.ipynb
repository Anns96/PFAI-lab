```python
class WaterJugProblem:
    def __init__(self, capacity_a, capacity_b, goal):
        self.capacity_a = capacity_a
        self.capacity_b = capacity_b
        self.goal = goal
        self.visited = set()
    def dfs(self):
        stack = []
        initial_state = (0, 0)
        stack.append((initial_state, []))
        while stack:
            current_state, path = stack.pop()
            self.visited.add(current_state)
            if current_state[0] == self.goal or current_state[1] == self.goal:
                return path
            next_states = self.get_next_states(current_state)
            for state, action in next_states:
                if state not in self.visited:
                    stack.append((state, path + [action]))
        return None
    def get_next_states(self, state):
        a, b = state
        next_states = []
        if a < self.capacity_a:
            next_states.append(((self.capacity_a, b), 'fill A'))
        if b < self.capacity_b:
            next_states.append(((a, self.capacity_b), 'fill B'))
        if a > 0:
            next_states.append(((0, b), 'empty A'))
        if b > 0:
            next_states.append(((a, 0), 'empty B'))
        if a > 0 and b < self.capacity_b:
            pour_amount = min(a, self.capacity_b - b)
            next_states.append(((a - pour_amount, b + pour_amount), 'pour A to B'))
        if b > 0 and a < self.capacity_a:
            pour_amount = min(b, self.capacity_a - a)
            next_states.append(((a + pour_amount, b - pour_amount), 'pour B to A'))
        return next_states
if __name__ == "__main__":
    jug_problem = WaterJugProblem(5, 3, 4)
    solution = jug_problem.dfs()

    if solution:
        print("solution found")
        for step in solution:
            print(step)
    else:
        print("no solution found")
```
