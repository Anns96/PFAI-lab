```python
from flask import Flask, render_template, request
from transformers import pipeline

app = Flask(__name__)

# load summarization pipeline (uses pretrained model)
summarizer = pipeline("summarization")

@app.route("/", methods=["GET", "POST"])
def index():
    summary = None
    original = None
    if request.method == "POST":
        original = request.form.get("text")
        if original and len(original.strip()) > 20:
            # generate summary (you can tune min_length / max_length)
            summary_obj = summarizer(original, max_length=130, min_length=30, do_sample=False)
            summary = summary_obj[0]['summary_text']
    return render_template("index.html", original=original, summary=summary)

if __name__ == "__main__":
    app.run(debug=True)
#html code

<!DOCTYPE html>
<html>
<head>
    <title>Text Summarizer</title>
    <style>
      body { font-family: Arial, sans-serif; padding: 40px; background: #f4f4f4; }
      textarea { width: 80%; height: 200px; padding: 10px; }
      button { padding: 10px 20px; font-size: 16px; margin-top: 10px; }
      .result { margin-top: 20px; padding: 15px; background: #fff; border-radius: 8px; box-shadow: 0 0 10px rgba(0,0,0,0.1); }
    </style>
</head>
<body>
  <h1>Text Summarizer</h1>
  <form method="POST">
    <textarea name="text" placeholder="Paste your text here...">{% if original %}{{ original }}{% endif %}</textarea><br>
    <button type="submit">Summarize</button>
  </form>

  {% if summary %}
    <div class="result">
      <h3>Summary:</h3>
      <p>{{ summary }}</p>
    </div>
  {% endif %}
</body>
</html>
```
